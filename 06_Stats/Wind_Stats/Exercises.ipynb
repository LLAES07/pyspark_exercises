### Step 6. Compute how many values are missing for each location over the entire record.  
#### They should be ignored in all calculations below. 

```python
data.show()
```

```python
from pyspark.sql import functions as F
from pyspark.sql.types import FloatType, DoubleType

exprs = []
for field in data.schema:             
    col = F.col(field.name)
    
    if isinstance(field.dataType, (FloatType, DoubleType)):
        cond = F.isnan(col) | col.isNull()
    else:
        cond = col.isNull()
    
    exprs.append(F.count(F.when(cond, field.name)).alias(field.name))

nan_counts = data.select(exprs)       # una sola pasada por el DataFrame
nan_counts.show(truncate=False)
```

```python
df.isna().sum(axis=0)
```

### Step 7. Compute how many non-missing values there are in total.

```python
from functools import reduce

total_missing = nan_counts.select(
    reduce(lambda a, b: a + b, (F.col(c) for c in nan_counts.columns))
).first()[0]

print(f"Número total de NaN/null en todo el DataFrame: {total_missing}")
```

```python
df.isna().sum().sum()
```

```python
from pyspark.sql import functions as F
from pyspark.sql.types import NumericType

num_cols = [f.name for f in data.schema if isinstance(f.dataType, NumericType)]

avg_exprs = [
    F.avg(                               # promedio por columna
        F.when(F.isnan(c) | F.col(c).isNull(), None)  # convierte NaN a null
    .otherwise(F.col(c))
    ).alias(c)
    for c in num_cols
]

row_avgs = data.agg(*avg_exprs).first()          # Row(...)

overall_mean = sum(row_avgs) / len(num_cols)

overall_mean
```
